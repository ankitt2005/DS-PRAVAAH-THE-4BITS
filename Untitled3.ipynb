```python
# CELL 2: THE FINAL APP (With Chat Logging)
import os
import json
import torch
import random
import time
import pandas as pd
import csv
from datetime import datetime
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain_huggingface import HuggingFacePipeline, HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.schema import Document
from langchain.memory import ConversationBufferMemory

# --- FILES ---
FILENAME = 'Conversational_Transcript_Dataset.json'
KB_CSV = 'causal_knowledge_base.csv'       # Saves the Analysis
CHAT_LOG_CSV = 'chat_session_logs.csv'     # Saves YOUR Questions & Answers
LIMIT = 100                                # Processing Limit

# ==========================================
# 1. GPU CHECK
# ==========================================
if not torch.cuda.is_available():
    print("âŒ WARNING: Running on CPU! Switch to T4 GPU for speed.")
else:
    print(f"âœ… GPU ACTIVE: {torch.cuda.get_device_name(0)}")

# ==========================================
# 2. DATA LOADER & PROCESSOR
# ==========================================
def load_and_process_data(llm):
    # CHECK IF ANALYSIS ALREADY EXISTS
    if os.path.exists(KB_CSV):
        print(f"\n[1/3] Found saved analysis in '{KB_CSV}'! Loading instantly...")
        df = pd.read_csv(KB_CSV)
        return df.to_dict('records')

    # LOAD RAW JSON
    print(f"\n[1/3] Loading Raw Data from {FILENAME}...")
    if not os.path.exists(FILENAME):
        print("âŒ ERROR: Please upload the JSON file!")
        return []

    try:
        data = json.load(open(FILENAME))['transcripts']
    except:
        raw = open(FILENAME, encoding='utf-8', errors='ignore').read()
        data = json.loads(raw[:raw.rfind('},')+1] + "\n  ]\n}")['transcripts']

    # RANDOM SAMPLE
    print(f"   Selecting {LIMIT} random conversations...")
    random.seed(42)
    random.shuffle(data)
    data = data[:LIMIT]

    # ANALYZE (CAUSAL EXTRACTION)
    print(f"[2/3] Analyzing {len(data)} items for Root Causes...")

    template = "<|system|>Analyst.</s>\n<|user|>Transcript: {transcript}\nOutcome: {outcome}\nFind Root Cause and Quote.\nFormat:\nCAUSE: [Reason]\nQUOTE: [Evidence]</s>\n<|assistant|>"
    prompt = PromptTemplate(template=template, input_variables=["outcome", "transcript"])
    chain = prompt | llm

    kb = []
    start = time.time()

    for i, t in enumerate(data):
        try:
            txt = "\n".join([f"{x['speaker']}: {x['text']}" for x in t['conversation']])
            res = chain.invoke({"outcome": t['intent'], "transcript": txt})

            if "CAUSE:" in res:
                parts = res.split("CAUSE:")[1].split("QUOTE:")
                cause = parts[0].strip()
                kb.append({"id": t['transcript_id'], "outcome": t['intent'], "cause": cause, "text": txt})

            # Save every 10 rows
            if i % 10 == 0 and i > 0:
                pd.DataFrame(kb).to_csv(KB_CSV, index=False)
                print(f"   Processed {i}/{LIMIT}...")
        except: pass

    pd.DataFrame(kb).to_csv(KB_CSV, index=False)
    print(f"âœ… ANALYSIS COMPLETE. Saved to {KB_CSV}")
    return kb

# ==========================================
# 3. MODEL LOADER
# ==========================================
def load_model():
    print("\n[Init] Loading AI Model...")
    model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype=torch.float16, device_map="auto")
    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=256, temperature=0.1)
    return HuggingFacePipeline(pipeline=pipe)

# ==========================================
# 4. CHAT SYSTEM (With Logging)
# ==========================================
def start_chat(llm, kb):
    print("\n[3/3] Building Search Engine...")
    docs = [Document(page_content=f"Outcome: {x['outcome']}\nCause: {x['cause']}", metadata=x) for x in kb]
    vectorstore = FAISS.from_documents(docs, HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))

    qa = RetrievalQA.from_chain_type(
        llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever(),
        chain_type_kwargs={
            "prompt": PromptTemplate(template="<|system|>Context: {context}\nHistory: {chat_history}\nQuestion: {question}</s><|assistant|>", input_variables=["context", "chat_history", "question"]),
            "memory": ConversationBufferMemory(memory_key="chat_history", input_key="question", output_key="result")
        }
    )

    # SETUP LOGGING
    if not os.path.exists(CHAT_LOG_CSV):
        with open(CHAT_LOG_CSV, 'w') as f:
            f.write("Timestamp,User_Query,AI_Response,Evidence_Used\n")

    print("\n=============================================")
    print(f"ðŸ¤– SYSTEM ONLINE. Logs saving to '{CHAT_LOG_CSV}'")
    print("=============================================")

    while True:
        q = input("\nUser: ")
        if q.lower() in ['exit', 'quit']: break

        try:
            # Get Answer
            response = qa.invoke({'query': q})
            ans = response['result']

            # Log to CSV
            with open(CHAT_LOG_CSV, 'a', newline='') as f:
                writer = csv.writer(f)
                writer.writerow([datetime.now(), q, ans, "See Knowledge Base"])

            print(f"AI: {ans}")
            print(f"   (Saved to {CHAT_LOG_CSV})")

        except: print("Error processing query.")

if __name__ == "__main__":
    llm = load_model()
    kb = load_and_process_data(llm)
    if kb:
        start_chat(llm, kb)
```
